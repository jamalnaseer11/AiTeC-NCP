```python
import numpy as np         #numpy is used for most of the mathematical (as np ables us to use it as a shortcut)
import matplotlib.pyplot as plt     # for charts and graphs
import pandas as pd



#IMPORTING THE DATASET
#1. ALWAYS REMEMBER TO CHANGE THE DIRECTORY OR PROVIDE THE FULL PATH


dataset = pd.read_csv('/kaggle/input/churn-modelling/Churn_Modelling.csv')       #stores all the data and columns into the array dataset
X= dataset.iloc[:, 3:13].values        # stores all the values of the columns except the last one (because last one is dependant )
y= dataset.iloc[:, 13].values            # stores all the values of the last one in y (dependant)




from sklearn.preprocessing import LabelEncoder,OneHotEncoder 
labelencoder_X= LabelEncoder()      #Labelencoder does simple encoding which gives values to the categories which is sometime not needed    // Onehotencoder      labelencoder_X = LabelEncoder()                                       #makes separate columns       
X[:, 1]= labelencoder_X.fit_transform(X[:, 1])                         # fits label encoding into first columns and then transforms it

from sklearn.preprocessing import LabelEncoder,OneHotEncoder 
labelencoder_X= LabelEncoder()      #Labelencoder does simple encoding which gives values to the categories which is sometime not needed    // Onehotencoder      labelencoder_X = LabelEncoder()                                       #makes separate columns       
X[:, 2]= labelencoder_X.fit_transform(X[:, 2])                         # fits label encoding into first columns and then transforms it



column = X[:, 1].reshape(-1, 1)
onehotencoder =OneHotEncoder()    
# inside brackets specifies that first column needs to be labeled
efc= onehotencoder.fit_transform(column).toarray()
X = np.hstack((X[:, 0:1], efc[:,1:3], X[:,2:]))     #dummy variable removed




from sklearn.model_selection import train_test_split
X_train, X_test , y_train , y_test = train_test_split(X, y , test_size= 0.2, random_state=0)     



# Feature Scaling
#this is rarely used because most of the libraires have built-in function of scaling
# it normalizes or standardizes the values that are too much out range mostly the new range of all values is -1 to 1

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.fit_transform(X_test)






# buildinf the ANN
import keras
from keras.models import Sequential
from keras.layers import Dense


#intitialize ANN
classifier = Sequential()

# maiking the first hidden layer
classifier.add(Dense(units= 6, kernel_initializer = 'uniform', activation= 'relu', input_dim=11))

# adding second layer
classifier.add(Dense(units= 6, kernel_initializer = 'uniform', activation= 'relu'))


#adding the output layer
classifier.add(Dense(units= 1, kernel_initializer = 'uniform', activation= 'sigmoid'))

#compiling sll the layers
classifier.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = ['accuracy'])


#fitting into train set
classifier.fit(X_train,y_train,batch_size=10, epochs= 100)

y_pred= classifier.predict(X_test)
y_pred=(y_pred>0.5)

print(y_pred)


from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test, y_pred)

print(cm)
```
