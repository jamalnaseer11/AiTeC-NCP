```python
import numpy as np         
import matplotlib.pyplot as plt     
import pandas as pd



dataset = pd.read_csv('/kaggle/input/salary-csv/Salary_Data.csv')       
X= dataset.iloc[:, :-1].values        
y= dataset.iloc[:, 1].values 




from sklearn.model_selection import train_test_split
X_train, X_test , y_train , y_test = train_test_split(X, y , test_size= 1/3, random_state=0)



#FITTING LINEAR REGRESSION INTO THE TRAINING SET
#linear regression is basically plottig a best fit line between the training sets dependent and independent values and then predicting the upcoming values
# the X_test set is the used to create predicted values of the values so that it can be compared with the test y values i.e y_test

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)                                   # a regressor have been made using the train dataset


#predicting the test set values
# you can also manually check with the oringial y_test value to see accuracy

y_pred= regressor.predict(X_test)                                  # array y_pred will give the predicted values


#PLOTTING GRAPHS FOR BETTER VISUALIZATIONS

plt.scatter(X_train, y_train, color = 'red')
plt.plot (X_train, regressor.predict(X_train), color = 'blue')
plt.title ('Salary vs Experience')
plt.xlabel('EXP')
plt.ylabel('salary')
plt.show()

# The above graph will show the train values scattered and along with the best fit line

# The below will show the predicted line or prediceted values and the y_test values (the predcited line will be samwe so the regressor.predict(X_train) will remain same)


plt.scatter(X_test, y_test, color = 'red')
plt.plot (X_train, regressor.predict(X_train), color = 'blue')
plt.title ('Salary vs Experience')
plt.xlabel('EXP')
plt.ylabel('salary')
plt.show()
```
